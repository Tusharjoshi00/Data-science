```python
x=df.drop("label",axis =1)
```

```python
##standard scaler ----normalise ----outliers se deal krne k liye
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_transformed = sc.fit_transform(x)
x_transformed.shape
```

```python
desc = pd.DataFrame(x_transformed).describe()
desc.head()
```

```python
df.describe()
```

```python
desc_formatted = desc.applymap(lambda x:f"{int(x):d}")
```

```python
from sklearn.cluster import KMeans
kmeans =KMeans()
kmeans.fit(x_transformed)
```

```python
label = kmeans.labels_
```

```python
df["cluster_8"] =label
```

```python
df["cluster_8"].value_counts()
```

```python
df[df["label"]=="rice"]["cluster_8"].value_counts()
```

```python
df[df["cluster_8"]== 1 ]["label"].value_counts()
```

```python
## met
from sklearn.metrics import silhouette_score
## range ==== -1 to 1
silhouette_score(x_transformed,label)
```

```python
kmeans.inertia_
```

```python
WCSS = {}
for i in range(5,51):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(x_transformed)
    WCSS[i] =kmeans.inertia_
```

```python
WCSS
```

```python
plt.figure(figsize = (20,8))
plt.plot(WCSS.keys(), WCSS.values(),marker ="o")
#add a line
plt.xticks(list(WCSS.keys()))
plt.grid(True)
plt.show()
```
